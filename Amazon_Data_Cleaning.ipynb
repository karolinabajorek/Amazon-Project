```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Reading the file
df = pd.read_csv('final_data.csv', delimiter = ',')
```

```python
# Checking if the data is correctly uploaded
df.head()
```

```python
# NULL values in Location and State columns - only for users from outside of the US
df.info()
```

```python
# Converting date-related columns to datetime
# date_columns = ['Date of Birth', 'Membership Start Date', 'Membership End Date']
# df[date_columns] = df[date_columns].apply(pd.to_datetime)
df['Date of Birth'] = pd.to_datetime(df['Date of Birth'])
df['Membership Start Date'] = pd.to_datetime(df['Membership Start Date'])
df['Membership End Date'] = pd.to_datetime(df['Membership End Date'])
```

```python
df.info()
```

```python
# Renaming the columns to a more convenient format
df.rename(columns = {
    'Location': 'location',
    'State': 'state',
    'Country': 'country',
    'User ID': 'user_id',
    'Name': 'name',
    'Email Address': 'email',
    'Username': 'username',
    'Date of Birth': 'date_of_birth',
    'Gender': 'gender',
    'Membership Start Date': 'membership_start_date',
    'Membership End Date': 'membership_end_date',
    'Subscription Plan': 'subscription_plan',
    'Payment Information': 'payment_information',
    'Renewal Status': 'reneval_status',
    'Usage Frequency': 'usage_frequncy',
    'Purchase History': 'purchase_history',
    'Favorite Genres': 'favorite_genres',
    'Devices Used': 'devices_used',
    'Engagement Metrics': 'engagement_metrics',
    'Feedback/Ratings': 'feedback_ratings',
    'Customer Support Interactions': 'customer_support_interactions'
    }, inplace=True)
```

```python
df.head()
```

```python
df.columns
```

```python
# Changing the order of columns
df = df[['user_id','location', 'state', 'country', 'name', 'email', 'username',
       'date_of_birth', 'gender', 'membership_start_date',
       'membership_end_date', 'subscription_plan',
       'reneval_status','payment_information', 'usage_frequncy', 'purchase_history',
       'favorite_genres', 'devices_used', 'engagement_metrics',
       'feedback_ratings', 'customer_support_interactions']]
```

```python
df.head()
```

```python
# Dropping irrelevant columns
df.drop(columns = ['location','name','state', 'email', 'username', 'engagement_metrics', 'favorite_genres', 'devices_used'], inplace=True)
```
